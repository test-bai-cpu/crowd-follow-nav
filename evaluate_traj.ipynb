```python
######## get navigation time #########
import os
import numpy as np

def get_navigation_time(trajectory):
    """
    Calculate the navigation time from the trajectory.
    The navigation time is defined as the time taken to reach the goal.
    """
    start_time = trajectory[0]['time']
    end_time = trajectory[-1]['time']
    dt = 0.1
    nav_time = (end_time - start_time) * dt

    return nav_time

def get_goal_reach_success(trajectory, goal, reach_threshold=0.1):
    """
    Determine if the goal was reached.
    Returns True if the goal was reached, False otherwise.
    """
    # goal is array-like, e.g., [0.5, 0.5]
    robot_pos = trajectory[-1]['robot_pos']
    goal_reached = np.linalg.norm(robot_pos - goal) < reach_threshold
    
    return goal_reached

def get_path_length(trajectory):
    """
    Calculate the path length from the trajectory.
    The path length is the sum of distances between consecutive positions.
    """
    robot_positions = np.array([step['robot_pos'] for step in trajectory])
    path_length = np.sum(np.linalg.norm(robot_positions[1:] - robot_positions[:-1], axis=1))

    return path_length

def get_minimum_pedestrian_distance(trajectory):
    """
    Calculate the minimum distance to any pedestrian in the trajectory.
    Returns the minimum distance.
    """
    
    min_dists = []

    for step in trajectory:
        robot_pos = step['robot_pos']             # shape: (2,)
        ped_pos = step['pedestrians_pos']         # shape: (N, 2)

        # Compute all distances from robot to pedestrians at this time
        dists = np.linalg.norm(ped_pos - robot_pos, axis=1)

        # Store the minimum distance at this time step
        min_dists.append(np.min(dists))

    # Get the minimum distance over the whole trajectory
    global_min_dist = np.min(min_dists)

    return global_min_dist


```
